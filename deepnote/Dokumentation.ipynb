## Basics Python
The most ***common function*** of python is print.
It works as follows:
Print(“Hello, World!”)

When you run the function it will print “Hello, World!”

Python uses indentation to define blocks of code. Example:
if x > 10:
	print(“x is greater than 10”)
else:
	print(“x is not greater than 10”)

Basic syntax rules:
-	Python statements are typically one per line.
-	If a statement is too long to fit on a single line, you can use a backslash to continue the statement on the next line
-	Comments are added to the code using the # symbol
(Example: # This is a comment )
-	Comments are ignored by the python interpreter and are used to provide explanations within the code
-	Python does not require semicolons to end statements. Each line represents a new statement

In python, a colon is to start a new block of code, such as in “if” statements, loops, and functions:
if condition:
	# this is a block of code
	Print(“Condition is true”)
Else:
	# another block of code
	Print(“Condition is false”)

### Variables and Datatypes
-	Variable names are case-sensitive, so myVar and myvar are treated as two different variables
-	Varbiable names can consist of letters, numbers, and underscores
-	Variable names cannot start with a number
-	Python has reserved words (if, else, elif, while, etc.) which cannot be used as variable names

***Numeric data types*** can be integer (int) and floating-point (float). An integer is a full number while a float is a decimal number.
Integers will be used as: x = 5, y = -10, and you can add addition, subtraction, multiplication, and division to the variable. Addition = x + y, subtraction = x – y, etc.
Floats work the same way, just with decimal numbers. X = 5.5, y = 2.0 -> addition = x + y, subtraction = x – y.

Integers can be converted to floats and the other way around. To do that, the float and integer need to be defined first. 
Integer_number = 10
Float_number = 3.13

To convert the following function needs to be used:
Convert_to_float = float(10.0)
Convert_to_int = int(3)

The ***text data type*** string (str) is used to represent text, characters, and strings. For that both single quotes (‘ ‘) or double quotes (“ “) can be used. 
Example: name = ‘Alice’ / name  = “Alice”

Strings can be combined by using the + operator:
First_name = ‘John’
Last_name = ‘Doe’

Full_name + ‘ ‘ + last_name = John Doe

String methods can also be used as follows:
Upper(): concerts a string into uppercase
Lower(): converts a string into lowercase
Strip(): removes leading and trailing whitespace
Replace(old, new): replaces occurrences of a substring with another
Split(delimiter): splits a string into a list of substrings based on a delimiter(Begrenzungszeichen)

A ***boolean*** is a variable that decides whether an information is True or False. 
you simply use x = true or x = false as a function.

You can change one data type into another with ***type conversion (also known as Casting).*** Python provides built-in functions for type conversion, allowing you to work with data in the format you need.
Python automatically performs some type convedrsions when necessary.
You can also perform type conversions by using built-in functions. The functions are:
-	Int(): converts a value to an integer
-	Float(): converts a value to a floating-point number
-	Str(): converts a value to a string
-	Bool(): converts a value to a Boolean

Numbers and strings can be combined:
Age = 30
Message = “I am” + str(age) + “years old.”
    # “I am 30 years old.”

***Conditional statements*** (if, elif(short for “else if”), else) are statements whether certain conditions are met or not.
The if statement allows you to execute a block of code if a condition is true.
if condition1: 
	# code to be executed if the condition1 is true
Elif is used to check additional conditions if the initial “if” condition is false.
elif condition2:
	# code to be executed if condition2 is true
The else statement is used to execute a block of code when none of the previous conditions are true. It doesn’t have a condition of its own but is associated with the preceding if or elif statements:
Else:
	# code to be executed if the condition is false
All of this can be combined:
x = 10
if x > 5:
	if x < 15:
		print(“x is between 5 and 15”)
	else:
		print(“x is greater than or equal to 15”)
else: 
	print(“x is not greater than 5”)

python has the following ***comparison operators:***
-	Equality operator “==”, it checks if two values are equal. It turns “True” if they are equal and “False” if they aren’t.
-	Inequality operator “!=”, it checks if two values are not equal. It turns “True” if they aren’t equal and “False” if they are.
-	Greater than operator “>”, it checks if the value on the left is greater than the value on the right. It turns “True” if the condition is met and “False” otherwise
-	Less than operator “<”, it checks if the value on the right is greater than the value on the left. It turns “True” if the condition is met and “False” otherwise
-	Greater than or equal to operator “>=”, it checks if the value on the left is greater than or equal to the value on the right. It turns “True” if the condition is met, “False” if otherwise.
-	Less than or equal to operator “<=”, it checks if the value on the left is less than or equal to the value on the right. It turns “True” if the condition is met, “False” if otherwise

The ***logical operators*** are “and”, “or”, “not”.
-	And: returns “true” if both conditions it connects are true, if at least one is false, it returns “false”
-	Or: returns “true” if at least one of the conditions it connects “True”, if both conditions are false, it returns “False”
-	Not: is used to reverse the logical value of a condition. If a condition is “true” the “not” operator makes it “False” and vice versa.

***Loops*** are defined by “for” and “while”
The for loop is use for iterating over a sequence (such as a list, tuple, string, or range) and executing a block of code for each item in the sequence:
for variable in sequence:
	# code to be exected for each item in the sequence

The while loop is used to repeatedly execute a block of code as long as a specified condition is true:
while condition:
	# code to executed while the condition is True

Loops can be controlled with the control statements:
-	Break: terminates the loop prematurely, even if the loop condition is still true
The break statement is commonly used to exit a loop when a certain condition is met
for item in sequence:
	if condition:
		break # terminate the loop

-	Continue: skips the rest of the current iretation and moves to the next iretation
It is useful when you want to bypass certain elements or conditions and continue with the next iretation
for item in sequence:
	if condition:
		continue # skip the rest of this iretation and continue with the next

-	Pass: acts as a placeholder, allowing you to write a loop with no content for testing or future implementation

you can assign default values to function parameters. These defaults are used if no value is provided for the corresponding parameter when calling the function. Example:
